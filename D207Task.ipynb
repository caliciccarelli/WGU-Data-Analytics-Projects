```python
#Begin by importing needed packages. They will allow for hypothesis testing and graphing.
import pandas as pd
import numpy as np
import statistics
from scipy import stats
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\Cali\Downloads\datamarch24\churn_clean.csv")
#Two-sample T-test
churn_incomes = df[df['Churn'] == 'Yes'] ['Income']
no_churn_incomes = df[df['Churn'] == 'No']['Income']
t_result= stats.ttest_ind(churn_incomes, no_churn_incomes)
print(t_result)
#is it significant? 
Alpha = 0.05
if (t_result[1] < Alpha):
    print("The mean income for those who chose to stop services and those who chose to stay are statistically different")
else: print("No significant differences between the mean incomes of those who renewed services and those who did not renew.")
#Distribution graph for Gender (catergorical)
gender_totals = df['Gender'].value_counts()
plt.bar(gender_totals.index, gender_totals.values)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
```

```python
# Distribution Graph for Contract variable (categorical):
contract_totals = df['Contract'].value_counts()
plt.pie(contract_totals.values, labels=contract_totals.index, autopct='%1.1f%%', startangle=90)
plt.title('Contract Type Distribution')
plt.axis('equal')
plt.show()
```
